```python
#Extracting job title.
def extract_job_title_from_result(soup): 
    jobs = []
    for div in soup.find_all(name="div", attrs={"class":"flex-row"}):
      for div2 in div.find_all(name="div", attrs={"class":"summary"}):
        for a in div2.find_all(name="h2", attrs={"class":"title"}):
          jobs.append(a.text.strip())
    return(jobs)
title = extract_job_title_from_result(soup)
len(title)
```

```python
#Extract Companies looking for employees.
def extract_company_from_result(soup): 
    companies = []
    for div in soup.find_all(name="div", attrs={"class":"flex-row"}):
      for div1 in div.find_all(name="div", attrs={"class":"summary"}):
        for div2 in div1.find_all(name="div", attrs={"class":"company"}):
          for company in div2.find_all(name="span", attrs={"class":"name"}):
            # if isinstance(company, NavigableString):
            #   continue
            if len(company) > 0:
              for b in company:
                companies.append(b.strip())
            else:
              sec_try = div.find_all(name="span", attrs={"class":"result-link-source"})
              for span in sec_try:
                  companies.append(span.text.strip())
    return(companies)
 
company = extract_company_from_result(soup)
len(company)
```

```python
#Extraction Location.
def extract_location_from_result(soup): 
    locations = []
    for div in soup.find_all(name="div", attrs={"class":"flex-row"}):
      for div1 in div.find_all(name="div", attrs={"class":"summary"}):
        for div2 in div1.find_all(name="div", attrs={"class":"location"}):
          for location in div2.find_all(name="span", attrs={"class":"name"}):
            if len(location) > 0:
              for b in location:
                locations.append(b.strip())
    return(locations)
location = extract_location_from_result(soup)
len(location)
```

```python
columns = ["[job title]", "[company]", "[location]"]

monster_df = pd.DataFrame(columns=columns)
#intermediate_df = pd.DataFrame(columns=columns)
#https://www.monster.com/jobs/search/?q=Data-Scientist&stpage=1&page=5
for start in range(2, 11):
    page1 = requests.get('https://www.monster.com/jobs/search/?q=Data-Scientist&stpage=1&page=' + str(start))
    time.sleep(1)
    soup = BeautifulSoup(page1.text, "lxml")
    intermediate_df = pd.DataFrame(columns=columns)
    titles = extract_job_title_from_result(soup)
    companies = extract_company_from_result(soup)
    locations = extract_location_from_result(soup)
    print(len(titles),len(companies),len(locations))
    intermediate_df[columns[0]] = titles
    intermediate_df[columns[1]] = companies
    intermediate_df[columns[2]] = locations
    monster_df.append(intermediate_df)        
#print(monster_df)
monster_df.to_csv("monster_data.csv", encoding='utf-8')
```
